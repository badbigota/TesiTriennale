```python
import scipy as sp
import scipy.signal as sps
import numpy as np
import matplotlib.pyplot as plt 
```

```python
signal=np.repeat((1,0,1,0,1,0),100)
href=sps.windows.gaussian(100,5)
```

```python
conv=sps.convolve(signal, href, mode='same')
conv2=np.convolve(signal, href)
```

```python
fig,ax=plt.subplots(2,2, sharex=True)
ax[0,0].plot(signal)
ax[0,0].set_title("Signal")
ax[1,0].plot(href)
ax[1,0].set_title("Gaussian")
ax[0,1].plot(conv)
ax[0,1].set_title("Convolved signal - Scipy")
ax[1,1].plot(conv2)
ax[1,1].set_title("Convolved signal - Numpy")
fig.tight_layout()
plt.show()
```

```python
a1=6
a2=12
b=0.9
c=0.35
d1=a1*b
d2=a2*b

x=np.linspace(0,25, 250)
y=((x/d1)**a1)*np.exp(-(x-d1)/b)-c*((x/d2)**a2)*np.exp(-(x-d2)/b)

plt.plot(x,y, label=r"$\left(\frac{t}{d_1}\right)^{a_1}\exp{\left(-\frac{t-d_1}{b_1}\right)}-c\left(\frac{t}{d_2}\right)^{a_2}\exp{\left(-\frac{t-d_2}{b_2}\right)}$")
plt.title("Hemodynamic response function")
plt.xlabel("Time $[s]$")
plt.ylabel("Amplitude A.U.")
plt.legend()
plt.show()
```
