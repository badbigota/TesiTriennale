```python
import numpy as np
from scipy import signal
import scipy.io.wavfile
import matplotlib.pyplot as plt
```

```python
# read ECG data from the WAV file
sampleRate, data = scipy.io.wavfile.read('ecg.wav')
times = np.arange(len(data))/sampleRate

#data=data[375:420]
sampleRate
```

```python
plt.specgram(data, Fs=sampleRate)
plt.show()
```

```python
numtaps=11
freqs=[2e-5,2e-4]
filt1=signal.firwin(numtaps, freqs, pass_zero=False)
filt2=signal.firwin(numtaps-6, freqs, pass_zero=False)
filt3=signal.firwin(numtaps+6, freqs, pass_zero=False)


filtered1=np.convolve(filt1, data, mode='valid')
filtered2=np.convolve(filt2, data, mode='valid')
filtered3=np.convolve(filt3, data, mode='valid')
#filtered4=np.convolve(np.convolve(filtered1, filt1, mode='valid'), data, mode='valid')
```

```python

plt.plot(data, 'r', label="Orig")
plt.plot(filtered1, 'b', label="Filt1")
plt.plot(filtered2, 'y', label="Filt2")
plt.plot(filtered3, 'g', label="Filt3")
#plt.plot(filtered4, 'm', label="Filt4")

plt.legend()
plt.show()
```
